## Handling outliers

```python
sns.boxplot(x='SepalWidthCm',data=lris);
```

## Here we see outliers arebelow 2.0 and above 4.0
_we know to remove the outliers by using IQR function i.e. q3-q1

```python

# IQR
Q1 = np.percentile(lris['SepalWidthCm'], 25,
                interpolation = 'midpoint')
  
Q3 = np.percentile(lris['SepalWidthCm'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
  
print("Old Shape: ", lris.shape)
  
# Upper bound
upper = np.where(lris['SepalWidthCm'] >= (Q3+1.5*IQR))
  
# Lower bound
lower = np.where(lris['SepalWidthCm'] <= (Q1-1.5*IQR))
  
# Removing the Outliers
lris.drop(upper[0], inplace = True)
lris.drop(lower[0], inplace = True)
  
print("New Shape: ", lris.shape)
  
sns.boxplot(x='SepalWidthCm', data=lris);
```
